```python
import torch
from usta_model import UstaModel
from usta_tokenizer import UstaTokenizer

u_tokenizer = UstaTokenizer("tokenizer.json")

prompt = "Artificial is intelligence is transforming the way people the interact with technology"
tokens = u_tokenizer.encode(prompt)

tokens
```

```python
torch.manual_seed(1)
u_model = UstaModel(vocab_size=len(u_tokenizer.vocab),embedding_dim = 4,num_heads = 2,context_length= 23,num_layers = 3)

sentence_meanings_with_attention_context = u_model(tokens)
sentence_meanings_with_attention_context.shape
```
