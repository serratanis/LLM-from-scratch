```python
from transformers import PreTrainedTokenizerFast
from tokenizers import normalizers

# 1. Hugging Face'teki mevcut tokenizer'ını çekiyoruz
tokenizer = PreTrainedTokenizerFast.from_pretrained("serratanis/hf_tokenizer")

# 2. Arka planda çalışan çekirdek tokenizer nesnesine ulaşıyoruz
backend_tokenizer = tokenizer.backend_tokenizer

# 3. Kural Setini Gömüyoruz (Normalizer)
# Artık tokenizer her metni gördüğünde ÖNCE küçük harfe çevirecek, SONRA nokta ve virgülleri silecektir.
backend_tokenizer.normalizer = normalizers.Sequence([
    normalizers.Lowercase(),
    normalizers.Replace(".", " "),
    normalizers.Replace(",", " ")
])

# 4. Bu yeni, akıllı tokenizer'ı Hugging Face repona üzerine yazarak güncelliyoruz
tokenizer.push_to_hub("serratanis/hf_tokenizer")

print("Tokenizer başarıyla güncellendi ve normalizer eklendi!")
```

```python
import plotly.graph_objects as go
import plotly.offline

def plot_dots(sentences_data,title):
    data=[
        go.Scatter3d(
            x = sentences_data["words"][:,0],
            y = sentences_data["words"][:,0],
            z = sentences_data["words"][:,0],
            mode="markers+text",
            marker=dict( 
                size = 6,
                color=sentences_data["color"]
            ),
            text=sentence_data["labels"],
            hoverinfo = "text",
        )   for sentence_data in sentences_data
    ]
    layout = go.Layout(
    scene=dict(
        xaxis_title="Sertlik",
        yaxis_title="Parlaklık",
        zaxis_title="Kırmızılık",
    ),
    title=title,
)
data = [
    go.Scatter3d(
        x=[0.1, 0.0, 0.30, 0.10],
        y=[0.4, 0.2, 0.06, 0.04],
        z=[0.0, 0.24, 0.03, 0.02],
        mode="markers+text",
        marker=dict(
            size=10,
            color="red",
        ),
        text=["artificial", "intelligence", "is", "transforming"],  # Virgül eklendi
        hoverinfo="text",
    )
]
layout = go.Layout(
    scene=dict(
        xaxis_title="Sertlik",
        yaxis_title="Parlaklık",
        zaxis_title="Kırmızılık",
    ),
    title="Sözlük V1",
)
fig = go.Figure(data=data, layout=layout)
plotly.offline.iplot(fig)
```

```python
import plotly.graph_objects as go
import plotly.offline

def plot_dots(sentences_data, title, dims=[0,1,2]):

    data = [
        go.Scatter3d(
            x=sentence_data["words"][:,dims[0]],
            y=sentence_data["words"][:,dims[1]],
            z=sentence_data["words"][:,dims [2]],
            mode="markers+text",
            marker=dict(
                size=10,
                color=sentence_data["color"]
            ),
            text=sentence_data["labels"],
            hoverinfo="text",
        )
        for sentence_data in sentences_data
    ]

    layout = go.Layout(
        scene=dict(
            xaxis_title="X",
            yaxis_title="Y",
            zaxis_title="Z",
        ),
        title=title,
    )

    fig = go.Figure(data=data, layout=layout)
    plotly.offline.iplot(fig)
```
