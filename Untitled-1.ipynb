```python
import json
with open("tokenizer.json","r") as f :
    vocab = json.load(f)
#this function gave us words ids 
def tokenize2(text):
    parts = text.split()
    ids = []
    for part in parts:
        if part in vocab:
            value =vocab[part]
        else :
            value = vocab["<unk>"]
        ids.append(value)
    return ids 

token_ids1
```

```python
reverse_vocab ={id: part for part , id in vocab.items()}
reverse_vocab
```

```python
def detokenize(ids):
    text = ""
    for id in ids:
        part  = reverse_vocab[id]
        text += part + " "
    text =text.strip()
    return text

detokenize(token_ids1)
```
